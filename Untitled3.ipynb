```python
Q1 = df['amount'].quantile(0.25)

Q3 = df['amount'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

outliers_iqr = df[(df['amount'] < lower) | (df['amount'] > upper)]

print("Number of IQR Outliers:", len(outliers_iqr))

outliers_iqr.head()
```

```python
from scipy import stats

df['ZScore'] = stats.zscore(df['amount'])

outliers_z = df[df['ZScore'].abs() > 3]

print("Number of Z-Score Outliers:", len(outliers_z))

outliers_z.head()
```

```python
fraud = df[df['isFraud']==1]['amount']

normal = df[df['isFraud']==0]['amount']

result = stats.ttest_ind(fraud, normal)

print(result)
```

```python
if result.pvalue < 0.05:
    print("Reject Null Hypothesis")
    print("Transaction amount significantly affects fraud.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant relationship.")
```
